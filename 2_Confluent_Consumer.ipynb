```python
from confluent_kafka import Consumer

conf = {
    'bootstrap.servers': 'pkc-7prvp.centralindia.azure.confluent.cloud:9092',
    'security.protocol': 'SASL_SSL',
    'sasl.mechanism': 'PLAIN',
    'sasl.username': 'user_name',
    'sasl.password': 'password',
    'group.id': 'jupyter-consumer-group',
    'auto.offset.reset': 'earliest' # Options: 'earliest' (all data) or 'latest' (new data only)
}

consumer = Consumer(conf)
print('mahesh')
```

```python
topic_name = "topic_policyDetails"
consumer.subscribe([topic_name])

try:
    while True:
        
        # poll(timeout) returns a message or None if no message is available
        msg = consumer.poll(timeout=1.0)
        print(msg)
        
        if msg is None:
            continue
        if msg.error():
            print(f"Consumer error: {msg.error()}")
            continue

        # Decode the message value (assuming it was sent as a JSON string)
        print(f"Received message: {msg.key().decode('utf-8')}")
        #print(f"Received message: {msg.value().decode('utf-8')}")
        #print(f"Received message: {msg.timestamp().decode('utf-8')}")
        

except KeyboardInterrupt:
    print("Stopped by user.")
finally:
    # Always close the consumer to clean up connections and commit final offsets
    consumer.close()
    print('stop')
```
